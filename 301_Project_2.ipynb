# Infant Mortality

Infant Mortality Rate by Maternal Race/Ethnicity for New York City, 2007-2016

Counts of infant deaths (age <1 year) are based on NYC death certificates. The rate is calculated using the counts of infant deaths as the numerator and the count of live births from NYC birth certificates as the denominator.

https://data.cityofnewyork.us/Health/Infant-Mortality/fcau-jc6k

*The 9 Columns in the dataset*:
1. YEAR: Year of death for infant in New York City.
2. Maternal Race or Ethnicity: Race/Ethnicity of the mother.
3. Infant Mortality Rate: Number of infant deaths under per 1,000 live births in the same calendar year in New York City .
4. Neonatal Mortality Rate: Number of infant deaths under 28 days of age per 1,000 live births in the same calendar year in New York City.
5. Postneonatal Mortality Rate: Number of infant deaths with age between 1 month and 1 year per 1,000 live births in the same calendar year in New York City.
6. Infant Deaths: Number of infant deaths in New York City.
7. Neonatal Infant Deaths: Number of infant deaths under 28 days of age in New York City.
8. Postneonatal Infant Deaths: Number of infant deaths with age between 1 month and 1 year in New York City.
9. Number of Live Births: Number of live newborns in New York City.

```python
import numpy as np
df.dropna(axis=0, how='any', inplace=True)
df.rename(columns={'Materal Race or Ethnicity':'Materal_Race_or_Ethnicity'}, inplace=True)
df.rename(columns={'Infant Mortality Rate':'Infant_Mortality_Rate'}, inplace=True)
df.rename(columns={'Neonatal Mortality Rate':'Neonatal_Mortality_Rate'}, inplace=True)
df.rename(columns={'Postneonatal Mortality Rate':'Postneonatal_Mortality_Rate'}, inplace=True)
df.rename(columns={'Infant Deaths':'Infant_Deaths'}, inplace=True)
df.rename(columns={'Neonatal Infant Deaths':'Neonatal_Infant_Deaths'}, inplace=True)
df.rename(columns={'Postneonatal Infant Deaths':'Postneonatal_Infant_Deaths'}, inplace=True)
df.rename(columns={'Number of Live Births':'Number_of_Live_Births'}, inplace=True)

df
```

# HISTOGRAM COUNT PLOTS

```python
import matplotlib.pyplot as plt
import seaborn as sns

selected_columns = ['Year', 'Materal_Race_or_Ethnicity', 'Infant_Mortality_Rate', 'Neonatal_Mortality_Rate', 'Postneonatal_Mortality_Rate', 'Infant_Deaths',
                     'Neonatal_Infant_Deaths', 'Postneonatal_Infant_Deaths', 'Number_of_Live_Births']
selected_df = df[selected_columns]
fig, ax = plt.subplots(ncols=3, nrows=3, figsize=(50, 20))
index = 0
ax = ax.flatten()
for col in selected_columns:
    if index >= len(ax):
        break
    g1 = sns.histplot(selected_df[col], ax=ax[index])
    g1.set(ylabel='count')
    ax[index].yaxis.label.set_size(20)
    ax[index].xaxis.label.set_size(20)
    ax[index].tick_params(axis="x", labelsize=20, rotation=90)
    ax[index].tick_params(axis="y", labelsize=15)
    index += 1
plt.tight_layout(pad=3.0, w_pad=7.0, h_pad=7.0)
plt.show()
```

# Train Model

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import OneHotEncoder, MinMaxScaler
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error

X = df[['Materal_Race_or_Ethnicity', 'Infant_Mortality_Rate', 'Neonatal_Mortality_Rate', 'Postneonatal_Mortality_Rate']]
y = df['Infant_Deaths']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=124)
encoder = OneHotEncoder(handle_unknown='ignore', sparse_output=False)
X_train_encoded = encoder.fit_transform(X_train)
X_test_encoded = encoder.transform(X_test)
```

```python
from sklearn.tree import DecisionTreeRegressor
model2 = DecisionTreeRegressor()
model2.fit(X_train_encoded, y_train)
predictions = model2.predict(X_test_encoded)
r2 = r2_score(y_test, predictions)
print(f'R-squared score: {r2}')
mse = mean_squared_error(y_test, predictions)
print(f'Mean squared error: {mse}')
```

```python
from sklearn.neighbors import KNeighborsRegressor
model3 = KNeighborsRegressor()
model3.fit(X_train_encoded, y_train)
predictions = model3.predict(X_test_encoded)
r2 = r2_score(y_test, predictions)
print(f'R-squared score on test data: {r2}')
mse = mean_squared_error(y_test, predictions)
print(f'Mean squared error: {mse}')

```
